## Site Description: Curlew National Grassland

The Curlew National Grassland consists of 47,600 acres of land located just South of Pocatello, Idaho (Beitia & Gunnell, 1986). It was established in 1960 to "improve soil and vegetation and promote sound agricultural practices" (Source 2). The land here was damaged by improper agrcultural techniques prior to the dustbowl. The establishment of it's National Grassland status initiated management by the US forest service, which has attempts to sustainably use the land for multiple purposes. Their management plan can be seen [here](https://www.fs.usda.gov/Internet/FSE_DOCUMENTS/stelprdb5238007.pdf) (Troyer & Resse). I chose this grassland primarily because of its small size compared to the other national grasslands, which made the data for this project more manageable.

## Site Description: Sheyenne National Grassland

Sheyenne National Grassland is located in Southeastern North Dakota, contains 70,180 acres of land, and is the only national grassland located in the tallgrass prarie region of the United States (Source 1).  The grassland is home to a number of sensitive species such as the Western Prarie Fringed Orchid, as well as many species of wildflowers and several unique ecosystems (Source 1, Source 2).
