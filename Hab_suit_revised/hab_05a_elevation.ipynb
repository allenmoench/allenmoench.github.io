```python
# Define a pattern to identify DEM tiles associated with Sheyenne National Grassland.
srtm_a_pattern = os.path.join(elevation_dir, 'N46*hgt.zip')
bounds_a = tuple(shey_gdf.total_bounds)
buffer = 0.25
xmin, ymin, xmax, ymax = bounds_a
bounds_buffer = (xmin-buffer, ymin-buffer, xmax+buffer, ymax+buffer)
if not glob(srtm_a_pattern):
    srtm_a_results = earthaccess.search_data(
        short_name = "SRTMGL1",
        bounding_box=bounds_buffer    
    )
    srtm_a_results = earthaccess.download(srtm_a_results, elevation_dir)
```

```python
# Print the DEM for Comanche natl. Grassland
srtm_a_da_list=[]
for srtm_a_path in glob(srtm_a_pattern):
    tile_da = rxr.open_rasterio(srtm_a_path, mask_and_scale=True).squeeze()
    cropped_da = tile_da.rio.clip_box(*bounds_buffer)
    srtm_a_da_list.append(cropped_da)
    
srtm_a_da = rxrmerge.merge_arrays(srtm_a_da_list)
srtm_a_da.plot(cmap='terrain')
shey_gdf.boundary.plot(ax=plt.gca(), color='black')
```
